```python
X = df_crime.iloc[:,1:]
X.head()
```

```python
# forming a group using clusturing....

from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean',linkage= 'complete')
Y = cluster.fit_predict(X)
```

```python
Y = pd.DataFrame(Y)
Y.value_counts()
```

```python

import matplotlib.pyplot as plt
#plt.figure(figsize=(10, 7))
plt.scatter(X.iloc[:,0], X.iloc[:,1], c=cluster.labels_, cmap='rainbow')
```

# **K MEAN CLUSTURING**

```python
from sklearn.cluster import KMeans
cluster = KMeans(n_init=20,n_clusters=4)
Y = cluster.fit_predict(X)
```

```python
Y=pd.DataFrame(Y)
Y.value_counts()
```

```python
import matplotlib.pyplot as plt
#plt.figure(figsize=(10, 7))
plt.scatter(X.iloc[:,0], X.iloc[:,1], c=cluster.labels_, cmap='rainbow')
```

```python
cluster.inertia_
```

```python
kinertias = []

for i in range(1,11):
    from sklearn.cluster import KMeans
    cluster = KMeans(n_init=20,n_clusters=i)
    cluster.fit_predict(X)
    kinertias.append(cluster.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.scatter(range(1,11),kinertias)
plt.plot(range(1,11),kinertias,color='red')
plt.show()
#from the k=4 the kinertia values are decreasing. It is best to have 4 clusters so that we can differentiate each cluster correctly

```
