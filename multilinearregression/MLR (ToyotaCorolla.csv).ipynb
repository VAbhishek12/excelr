# 1. Using Linear Model...

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("ToyotaCorolla.csv", encoding='latin1') 
df.shape

# we  have used "latin1" because 'UnicodeDecodeError' is occuring.
# UnicodeDecodeError is occur when pandas encounters a character 
# in the CSV file that it cannot decode using the specified encoding  
```

```python
df.head()
```

```python
df.corr() 
# finding the correlation of all the variables
#correaltion will help use to chose which variable should be used first 
```

```python
# Corolla<-Corolla[c("Price","Age_08_04","KM","HP","cc","Doors","Gears","Quarterly_Tax","Weight")]
Req_col = ["Price", "Age_08_04", "KM", "HP", "cc", "Doors", "Gears", "Quarterly_Tax", "Weight"]
df[Req_col].corr()
#finding the coreration of required variables as mention in problem statement 
# for that we form an array name as Req-col and assigned variables in it.
```

```python
y = df["Price"] # fixed varaible we need to predict
```

```python
x = df[["Age_08_04"]] # changes variable
# B0 +B1Age_08_04             where B = beta 
```

```python
# fitting the model..
from sklearn.linear_model import LinearRegression 
LR = LinearRegression()
LR.fit(x,y)
y_pred = LR.predict(x)

# finding the mean square and rsquare values 
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y,y_pred)
#print("mean squared error:",mse.round(2))
RMSE = np.sqrt(mse)
print("Root mean squared error:",RMSE.round(2))
r2 = r2_score(y,y_pred)
print("r square:",r2.round(2))
```

# 2. Stats Model..

```python
# we imported data set in above model so we directly imported stats model here..
import statsmodels.formula.api as smf
model = smf.ols('Price ~ Age_08_04 + KM + HP + cc + Doors + Gears + Quarterly_Tax + Weight',data = df).fit()
model.summary()
```

```python
print("Mean square error:",np.mean(model.resid ** 2))
```

```python
model.resid.hist() #histogram..
```

```python
#Resednitail analysis
#Test for normality 

import matplotlib.pyplot as plt
import statsmodels.api as sm
qqplot = sm.qqplot(model.resid,line = 'q')
plt.title('Normal Q-Q plot of residule')
plt.show()
```

```python
# Residual plot
import matplotlib.pyplot as plt
plt.scatter(model.fittedvalues,model.resid)
plt.title("Residual Plot")
plt.xlabel("Fitted Values")
plt.ylabel("residual Values")
plt.show()
```

```python
# checking the model influence 
model_influence = model.get_influence()
model_influence
(cooks,pvalue) = model_influence.cooks_distance
cooks = pd.DataFrame(cooks)
cooks[0].describe()
```

```python
# PLot the influence values..

fig = plt.subplots(figsize = (20,7))
plt.stem(np.arange(len(df)),np.round(cooks[0],3))
plt.xlabel("Row index")
plt.ylabel("Cooks Distance")
plt.show()
```
